```python
url = 'https://es.wikipedia.org/wiki/Fortune_Global_500'
html=requests.get(url).content
soup = bs(html,"html.parser")
table =soup.find_all('table')
good=table[-2]
good
```

```python
data=[]


for e in good.find_all('tr'):
    fila=[f for f in e.find_all('td')]
    if len(fila)>0:
        pais={'Ranking':str(fila[0].text),
              'Ciudad':fila[1].find('a').text.strip(),
              'Pais':fila[2].find('a').text.strip(),
              'Global companies':str(fila[3].text),
              'Ingreso':str(fila[4].text)}
        data.append(pais)
df=pd.DataFrame(data)
```

```python
df
```

```python
df.Ingreso = df.Ingreso.str.replace('\n', '')
df.Ingreso = df.Ingreso.str.replace(',', '')
df
```

```python
df['Ingreso'] = df['Ingreso'].apply(pd.to_numeric)
df['Global companies'] = df['Global companies'].apply(pd.to_numeric)
df['AvgRevenueXCompanie'] = df.Ingreso / df['Global companies']
df
```

```python
df.index = df.Ranking
df.drop(df.Ranking)
df
#Se puede observar el ingreso promedio de las compañias en el ranking 10 de las ciudades con mayor ingreso
#En la tabla se puede observar que a pesar de Muinch estar en el ranking 8, cada una de sus compañias tienen un ingreso promedio mayor que al de las demas, seguido de Londres que esta en el 5 lugar
```

```python
df2 = pd.read_json('https://www.thecocktaildb.com/api/json/v1/1/filter.php?a=Alcoholic')
for e in df2.drinks:
    print(e)
```

```python
columnas =[]
for items in df2.drinks:
    for key in items:
        if key not in columnas:
            columnas.append(key)
```

```python
strDrink=[]
strDrinkThumb=[]
idDrink =[]
for item in df2.drinks:
    if item['strDrink'] not in strDrink:   
        strDrink.append(item['strDrink'])
    if item['strDrinkThumb'] not in strDrinkThumb:   
        strDrinkThumb.append(item['strDrinkThumb'])
    if item['idDrink'] not in idDrink:   
        idDrink.append(item['idDrink'])
```

```python
len(idDrink)
```

```python
data = {columnas[0]:strDrink, columnas[1]:strDrinkThumb, columnas[2]:idDrink}
newdf = pd.DataFrame(data)
newdf.rename(columns={'strDrink':'DrinkName', 'strDrinkThumb':'Link', 'idDrink':'IdDrink'}, inplace=True)
newdf.head(10)
```
